## Exploratory Data Analysis

```python
df.columns
```

**Size**

```python
ax = sns.countplot(x='Size',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
#for loop is used to mention the count
```

**Note : From above graph we can say that most of the people buy M-Size**

### Group By

**The groupby() function in pandas is used to group data based on one or more columns in DataFrame**

```python
S_Qty = df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(x='Size',y='Qty',data=S_Qty)
```

**Note : From above graph we can see that most of the Qty buys are M-Size in the Sales**

## Courier Status

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(data=df, x='Courier Status', hue='Status')
plt.show()
```

**Note : The majority of the orders are successfully shipped and delivered to buyers, indicating an efficient shipping process with relatively few incidents of cancellations or returns.**



```python
#histogram
df['Size'].hist()
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data, bins=30, edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

**Note : From above graph we can see that most of the buyers are Buying T-shirts**

```python
#Checking B2B data by using piechart
B2B_check = df['B2B'].value_counts()
#plot the pie chart
plt.pie(B2B_check, labels=B2B_check.index, autopct='%1.1f%%')
plt.show()
```

**Note : From above chart we can see that max i.e, 99.2% of buyers are retailers and 0.8% are B2B buyers.**

```python
#Prepare data from scatter plot
x_data = df['Category']
y_data = df['Size']

#plot the scatter
plt.scatter(x_data , y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
#Plot count of cities by states
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of States')
plt.xticks(rotation = 90)
plt.show()
```

```python
#Top 10 states
top_10_states = df['ship-state'].value_counts().head(10)
#plot count of cities by states
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_states.index)], x = 'ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation = 45)
plt.show()
```

*Note : From above graph we can see that most of the buyers are from Maharastra state*
